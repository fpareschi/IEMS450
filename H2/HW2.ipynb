```python
import os 
import numpy as np
import sys

#dirpath = "C:/Users/franc/Google Drive/Northwestern/Winter - 2021/Optimization - 450-2/code"
#sys.path.append(dirpath)

import HW2_functions as fcts
```

```python
# Set key parameters to use across runs

tol = 1e-6
max_it = 10000

alpha_bar = 1
rho = 0.5
c = 10e-4

LSinst = fcts.LinearSearch(rho, c, alpha_bar, 2)
```

```python
x_0 = np.array([1.2, 1.0])
LSinst.main_loop(x_0, tol, max_it, method='Steepest')
LSinst.main_loop(x_0, tol, max_it, method='Newton')
LSinst.main_loop(x_0, tol, max_it, method='BFGS')
```

Observe that all algorithms converge to the solution. Nevertheless, the steepest descend method is notoriously slower, and it takes more than 200 iteration to conver. On the other hand the Newton and BFGS method converges in less than 20 iterations. It should be noted that for both Newton and BFGS we usually do only 1 function evaluation, per iteration. This theoretically expected, at least locally. 

```python
x_02 = np.array([-1.2, 1.0])
LSinst.main_loop(x_02, tol, max_it, method='Steepest')
LSinst.main_loop(x_02, tol, max_it, method='Newton')
LSinst.main_loop(x_02, tol, max_it, method='BFGS')
```

It is clearly harder to converge to the minimum from the new initial point. I'd like to highlight the fact that at the begining of the algorithm the number of function evaluations in the Newton and BFGS is quite hard, which points out that the quadratic model is not all that accurate in that region. 

```python
N = 100
x_0n =-np.ones(100)

LSinst = fcts.LinearSearch(rho, c, alpha_bar, N)

LSinst.main_loop(x_0n, tol, max_it, method='Steepest')
LSinst.main_loop(x_0n, tol, max_it, method='Newton')
LSinst.main_loop(x_0n, tol, max_it, method='BFGS')


```

It is clear from the outcomes, that finding the minimum of the extended Rosenbrock function is harder, which should not come as a surprise as the dimensionality of the problem increases significantly. In particular, note that the Steepest descend algorithm does not converge even after 15000 iterations. Nevertheless, if more iterations are allowed, the algorithm is expected to converge.
